Quadratic deviation depending of each language is too disparate to extrapolate. We're going to get rid of languages with too few occurence...

```python
df.groupby('language_code').language_code.count() > 100
```

```python
lang_to_keep = ['en-GB','en-US','eng','fre','spa']
df_ml = df[df.language_code.isin(lang_to_keep)]
df_ml.language_code.unique()
```

```python
X = df_ml.drop(columns='average_rating')
y = df_ml['average_rating']
```

```python
encoder = TargetEncoder()
X['author1_encoded'] = encoder.fit_transform(X['author1'], y)
X['author2_encoded'] = encoder.fit_transform(X['authors2'], y)
X['author3+_encoded'] = encoder.fit_transform(X['authors3+'], y)
X['title'] = encoder.fit_transform(X['title'], y)
```

```python
X['publisher_encoded'] = encoder.fit_transform(X['publisher'], y)
```

```python
X = X.drop('publisher', axis=1)
```

```python
X.info()
```

```python
X = X.drop(['author1','authors2','authors3+'], axis=1)
```

```python
X = pd.get_dummies(X, columns=['language_code'])
```

```python
X.info()
```

```python
#séparation pour le train et le test
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.9, test_size=.1, random_state=42)
```

```python
# En utilisant le modèle de regression forêt aléatoire
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error


model = RandomForestRegressor()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Calcul du MSE
mse = mean_squared_error(y_test, y_pred)
print(f"MSE: {mse}")
```
